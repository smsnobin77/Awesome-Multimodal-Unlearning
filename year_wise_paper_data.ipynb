```python
import matplotlib.pyplot as plt
import numpy as np

# Data
years = ['2022', '2023', '2024', '2025']
vlms = [2, 5, 22, 26]
dms = [1, 6, 32, 17]

x = np.arange(len(years))
width = 0.35

# Plot
plt.figure()
plt.bar(x - width/2, vlms, width, label='VLMs', color='#deba77')
plt.bar(x + width/2, dms, width, label='DMs', color='#7f77c8')

plt.xticks(x, years)
plt.xlabel('Year')
plt.ylabel('Number of Papers')
plt.title('VLMs vs DMs Papers by Year')
plt.legend(loc='upper left')

# save_path = "../vlm_dms_bar_chart.png"
# plt.savefig(save_path, dpi=600, bbox_inches="tight")

# print(f"Saved to: {save_path}")

plt.show()
```
